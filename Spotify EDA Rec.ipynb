```python
tracks['release_date'] = pd.to_datetime(tracks['release_date'],format = '%Y/%m/%d')
```

```python
tracks.set_index('release_date',inplace = True)
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x : round(x/1000))
```

```python
tracks.drop('duration_ms',axis = 1,inplace = True)
```

```python
track = tracks.drop(['key','explicit','mode'],inplace = True,axis = 1)
```

```python
corr_df = tracks.corr(method = 'pearson')
```

```python
plt.figure(figsize=(10,6))
heatmap = sns.heatmap(corr_df,annot = True,fmt = '0.1g',vmin = -1,vmax = 1
                      ,cmap = 'viridis',linewidths = 0.7,linecolor ='black',annot_kws = {'size':10})
heatmap.set_title('Correlation HeatMap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
plt.show()
```

```python
sample_df = tracks.sample(int(0.004*len(tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y= 'loudness',x = 'energy',color = 'c').set(title = 'Loudness vs Energy Correlation')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y= 'popularity',x = 'acousticness',color = 'g').set(title = 'Popularity vs Acousticness Correlation')
plt.show()
```

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
plt.figure(figsize=(8,4))
sns.set_style(style = 'whitegrid')
sns.displot(years,discrete = True,aspect = 2,height = 5,kind = 'hist',color = 'r').set(title = 'Numbers of song per year')
plt.show()
```

```python
plt.figure(figsize = (18,7))
sns.barplot(x = years,y = tracks.duration,errwidth = False).set(title = 'Years vs Duration')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize = (10,5))
sns.lineplot(x = years, y = tracks.duration).set(title = 'Year vs Duration')
plt.xticks(rotation = 60)
plt.show()
```

```python
genre_df = pd.read_csv('SpotifyFeatures.csv')
```

```python
genre_df.head()
```

```python
plt.figure(figsize = (10,6))
plt.title('Duration of Song in Different Genres')
sns.color_palette('rocket',as_cmap = True)
sns.barplot(x = 'duration_ms',y = 'genre',data = genre_df)
plt.xticks(rotation = 60)
plt.xlabel('Duration in milli Seconds')
plt.ylabel('Genres')
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_style(style = 'darkgrid')
famous = genre_df.sort_values('popularity',ascending = False).head(10)
sns.barplot(x = 'popularity',y = 'genre',data = famous).set(title = 'Top 5 Genres by Popularity')
plt.xticks(rotation = 60)
plt.show()
```
